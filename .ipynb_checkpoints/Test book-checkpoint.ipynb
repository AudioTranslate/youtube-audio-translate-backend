```python
def sythesize_segments(segments):
    final_audio = np.array([])
    prev_end = float('inf')
    for seg in segments:
        expected_duration = seg['end'] - seg['start']
        _break = seg['start'] - prev_end
        res = np.array([])
        if _break > 0:
            residual = np.zeros((_break * sr,))
            res = np.concatenate((res, residual))

        audioframe = np.array(tts.tts(seg['text'],  speaker=speaker, language=lang))
        output_duration = len(audioframe) // sr
        time_diff = abs(output_duration - expected_duration)

        if time > 0:
            if output_duration < expected_duration;
                audioframe = np.concatenate((np.zeros((time_diff * sr, )), audioframe))
            else:
                rate = expected_duration / output_duration
                audioframe = tsm.wsola(audioframe, rate)
            print(f'New rate = {len(audioframe) // sr}')
        res = np.concatenate((res, audioframe))
        final_audio = np.concatenate((final_audio, res))
        prev_end = seg['end']
    return final_audio
```
